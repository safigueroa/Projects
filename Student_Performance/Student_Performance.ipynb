```python
import numpy as np
np.set_printoptions(suppress=True)
import matplotlib.pyplot as plt
import seaborn as sns
np.seterr(divide='ignore', invalid='ignore')
%matplotlib inline
```

```python
studentMat = np.genfromtxt('studentMat.txt',filling_values=np.nan,dtype=float,skip_header=1)
```

```python
studentPor = np.genfromtxt('studentPor.txt',filling_values=np.nan,dtype=float,skip_header=1)
```

```python
clmNames = ["sex","age","address","famsize", "Pstatus", "Medu", "Fedu", "traveltime", "studytime", "failures", "activities", "higherEdu", "internet", "romantic", "famRel", "goOut", "absences", "G1", "G2", "G3"]
```

```python
studentMat[:10,:]
```

```python
studentPor[:10,:19]
```

```python
# Assign predictors to Xnames
feature_cols = ["sex","age","address","famsize", "Pstatus", "Medu", "Fedu", "traveltime", "studytime", "failures", "activities", "higherEdu", "internet", "romantic", "famRel", "goOut", "absences", "G1", "G2"]
# Assign Price to Target T
Target_col = ["G3"]
```

```python
X = studentPor[:, :19]
```

```python
T = studentPor[:, 19]
```

```python
#Plot regression plot each predictor vs. Target
plt.figure(figsize=(30,20))
for c in range(X.shape[1]):
    plt.subplot(5,5, c+1)
    sns.regplot(X[:,c], T, 'o')
    plt.ylabel(Target_col)
    plt.xlabel(feature_cols[c])
```

```python
print("Summary Statistics for Studytime: \n\tAvg:" , np.mean(studentPor[:,8]) , "\n\tStdev:", np.std(studentPor[:,8]) ,
       "\n\tMin:" , np.min(studentPor[:,8]), "\n\tMax:" , np.max(studentPor[:,8]))
```

```python
print("Summary Statistics for Address: \n\tAvg:" , np.mean(studentPor[:,2]) , "\n\tStdev:", np.std(studentPor[:,2]) ,
       "\n\tMin:" , np.min(studentPor[:,2]), "\n\tMax:" , np.max(studentPor[:,2]))
```

```python
print("Summary Statistics for Absences: \n\tAvg:" , np.mean(studentPor[:,16]) , "\n\tStdev:", np.std(studentPor[:,16]) ,
       "\n\tMin:" , np.min(studentPor[:,16]), "\n\tMax:" , np.max(studentPor[:,16]))
```

```python
print("Summary Statistics for Activities: \n\tAvg:" , np.mean(studentPor[:,10]) , "\n\tStdev:", np.std(studentPor[:,10]) ,
       "\n\tMin:" , np.min(studentPor[:,10]), "\n\tMax:" , np.max(studentPor[:,10]))
```

```python
#Confusion Matrix
corr = np.corrcoef(studentPor[:,:].T)

corr_mat = np.triu(corr)
```

```python
plt.figure(figsize = (20,10))
# plot the heatmap and annotation on it
sns.heatmap(corr, xticklabels=clmNames, yticklabels=clmNames, annot=True, cmap='coolwarm',cbar_kws={'label': 'Correlation'},mask=corr_mat)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Split data into train/test
xTrain, xTest, yTrain, yTest = train_test_split(X, T, test_size = 0.3, random_state = 0)
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(xTrain , yTrain)
```

```python
#Make predictions using test data
price_y_pred = reg.predict(xTest)
```

```python
plt.scatter(price_y_pred , (yTest - price_y_pred) , color='red')
plt.title("Residuals for Regression Model")
plt.xlabel("Predicted Value \"Price\" ")
plt.ylabel("Residuals")
plt.axhline(y=0.0, color='blue', linestyle='-')
plt.show()
```

```python
#Accuracy of test for studentPor
print('Test Score:', reg.score(xTest,yTest))
```

```python
from sklearn import metrics
```

```python
#Two very important metrics for multi-variable regression
print('MSE :', metrics.mean_squared_error(yTest,price_y_pred))

print('RMSE :', np.sqrt(metrics.mean_squared_error(yTest,price_y_pred)))
```
